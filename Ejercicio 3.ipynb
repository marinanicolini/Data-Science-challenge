```python
%%time
import pandas as pd

def preproc(item):
    return item.lower().replace('-',' ').replace('+',' ').replace('ê','e').replace('é','e').replace('ã','a').replace('â','a').replace('ç','c')


def score(item1, item2):
    item1, item2 = set(item1.split()), set(item2.split())
    x = len(item1 & item2)
    score = 2 * x / (len(item1) + len(item2))
    return score


items = pd.read_csv('items_titles.csv')
items_test = pd.read_csv('items_titles_test.csv')

#Tomamos un sample de 1000 productos para analizar el algoritmo en un tiempo razonable.
df = items_test.sample(1000)
#df = items_test 

df = df.rename(columns={'ITE_ITEM_TITLE':'title'})
df['title'] = df['title'].map(lambda x: preproc(x))

cross = df.merge(df, how='cross')

cross['score'] = cross.apply(lambda x: score(x['title_x'],x['title_y']),axis=1)

cross = cross.sort_values('score',ascending=False)

cross.head()
```
