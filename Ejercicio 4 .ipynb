```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from prophet import Prophet
from prophet.diagnostics import cross_validation
from prophet.diagnostics import performance_metrics
from prophet.plot import plot_cross_validation_metric

def predictions(df, categoria):
    df_cat = df[df['CATEGORY']==categoria].drop(['CATEGORY'], axis=1).sort_values('DATE').reset_index(drop=True)
    df_cat.rename(columns={'DATE':'ds','UNITS_SOLD':'y'}, inplace=True)
    
    model = Prophet(daily_seasonality=True)
    model.fit(df_cat)
    
    future_dates = model.make_future_dataframe(periods=21, freq = "D")
    
    prediction = model.predict(future_dates)

    return(prediction[['ds','yhat']].rename(columns={'ds': 'DATE', 'yhat': 'UNITS_SOLD'})) 
```

# Generamos las predicciones del .csv dado

```python
data = pd.read_csv('series.csv').sort_values('DATE') 
data
```
